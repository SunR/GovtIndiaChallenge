## First, we use logistic regression to predict a binary variable, whether the candidate had significant Computer Science skill. Then we repeat for each type of engineering listed to find the qualities that most contribute to skill in these areas. 

```python
import pandas as pd
path = 'C:/WinPython-64bit-3.5.3.0Qt5/scripts/Summer17-week1/datachallenge_cods2016/train.xlsx'
frame = pd.read_excel(path)
frame
```

## Use ComputerScience skill (or any other skill) as label, and everything else one-hot encoded as input. 
Pay special attention to gender, college tier, and personality types

### This way the govt will know what contributes to high-quality engineers. Including good teaching?

```python
frame.columns
```

```python
frame_cs = frame.drop(['ID', 'Salary', 'DOJ', 'DOL', 'Designation', 'JobCity', 'DOB', 'Domain', 'Specialization',
                       'ComputerProgramming', 'ElectronicsAndSemicon', 'ComputerScience', 'MechanicalEngg', 
                       'ElectricalEngg', 'TelecomEngg', 'CivilEngg', 'Degree'], axis=1)
```

```python
frame_cs
```

```python
import pandas as pd

frame_cs_transformed = pd.get_dummies(frame_cs)
print ([i for i in frame_cs_transformed.columns])
```

```python
frame_cs_transformed['years_since_grad'] = [2017-i for i in frame_cs_transformed['GraduationYear']]
frame_cs_transformed.drop(['GraduationYear', 'CollegeID', 'CollegeCityID', '12graduation'], axis=1, inplace=True)
```

```python
frame_cs_transformed
```

```python
labels_cs = frame['ComputerScience'].map(lambda r: r > 0)
labels_cs
```

```python
labels_cs.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
clf = LogisticRegression()
X_train, X_test, y_train, y_test = train_test_split(frame_cs_transformed, labels_cs)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import roc_auc_score
import operator
y_pred = pd.DataFrame(clf.predict_proba(X_test))
#print y_pred.iloc[0]
roc_auc = roc_auc_score(y_test, y_pred[1])
print (roc_auc)
coefs = {}
for i in clf.coef_:
    for j in range(len(i)):
        coefs[frame_cs_transformed.columns[j]] = i[j]
top_coefs = dict(sorted(coefs.items(), key=operator.itemgetter(1), reverse=True)[:10])
bottom_coefs = dict(sorted(coefs.items(), key=operator.itemgetter(1))[:10])
#print ("Coefficients: ", coefs)
print
print ("Top 10: ", top_coefs)
print
print ("Bottom 10: ", bottom_coefs)
```

## What we see above is the area under the ROC curve for this logistic regression model -- 0.82, which is quite good. We also see the top 20 coefficients affecting the model in either direction
